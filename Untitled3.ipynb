```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy.stats import t

import warnings
warnings.filterwarnings('ignore')
```

计算top20的航班数量的origin，在疫情前中后的载客量占总载客量的比例

```python
filtered_data=pd.read_csv("filtered_com_major_final.csv")
filtered_data
```

```python
columns_to_drop = ['Unnamed: 0','destination','Unnamed: 0','firstseen','lastseen','Olatitude','Olongitude','Dlatitude','Dlongitude']
filtered_data = filtered_data.drop(columns=columns_to_drop)
filtered_data
```

```python
filtered_data['DateTime'] = pd.to_datetime(filtered_data['day'])
filtered_data['Date'] = filtered_data['DateTime'].dt.date
```

```python
filtered_data
```

```python
PassengerNumber=pd.read_csv("PassengerNumber.csv")
PassengerNumber
```

```python
#根据typecode合并
filtered_data_total = pd.merge(filtered_data, PassengerNumber, on='typecode', how='inner')
filtered_data_total_sort = filtered_data_total.sort_values(by='typecode')
filtered_data_total_sort
```

```python
filtered_data_clean = filtered_data_total_sort.dropna(subset=['Passenger Capacity'])
filtered_data_clean
```

```python
filtered_data_beforeCovid = filtered_data_clean[(filtered_data_clean['Date'] >= pd.to_datetime('2019-01-01').date()) & (filtered_data_clean['Date'] <= pd.to_datetime('2019-12-31').date())].reset_index(drop=True)
filtered_data_beforeCovid
```

```python
filtered_data_duringCovid = filtered_data_clean[(filtered_data_clean['Date'] >= pd.to_datetime('2020-01-01').date()) & (filtered_data_clean['Date'] <= pd.to_datetime('2021-12-31').date())].reset_index(drop=True)
filtered_data_duringCovid
```

```python
filtered_data_afterCovid = filtered_data_clean[(filtered_data_clean['Date'] >= pd.to_datetime('2022-01-01').date()) & (filtered_data_clean['Date'] <= pd.to_datetime('2022-12-31').date())].reset_index(drop=True)
filtered_data_afterCovid
```

# 计算疫情前

```python
counts_beforeCovid = filtered_data_beforeCovid['origin'].value_counts()
result_beforeCovid = counts_beforeCovid.reset_index()
result_beforeCovid.columns = ['origin', 'Count']
result_beforeCovid.head(20)
```

```python
origin_values_beforeCovid = result_beforeCovid['origin'].unique()
top_20_beforeCovid = origin_values_beforeCovid[:20]
top_20_beforeCovid
```

```python
filtered_data_top20_beforeCovid = filtered_data_beforeCovid[filtered_data_beforeCovid['origin'].isin(top_20_beforeCovid)].reset_index(drop=True)
filtered_data_top20_beforeCovid
```

```python
total_passenger_beforeCovid = filtered_data_beforeCovid['Before Covid'].count()
total_passenger_beforeCovid
```

```python
top20_passenger_beforeCovid = filtered_data_top20_beforeCovid['Before Covid'].count()
top20_passenger_beforeCovid
```

```python
top20_passenger_beforeCovid_percentage = (top20_passenger_beforeCovid/total_passenger_beforeCovid)*100
top20_passenger_beforeCovid_percentage
```

# 计算疫情中

```python
counts_duringCovid = filtered_data_duringCovid['origin'].value_counts()
result_duringCovid = counts_duringCovid.reset_index()
result_duringCovid.columns = ['origin', 'Count']
result_duringCovid.head(20)
```

```python
origin_values_duringCovid = result_duringCovid['origin'].unique()
top_20_duringCovid = origin_values_duringCovid[:20]
top_20_duringCovid
```

```python
filtered_data_top20_duringCovid = filtered_data_duringCovid[filtered_data_duringCovid['origin'].isin(top_20_duringCovid)].reset_index(drop=True)
filtered_data_top20_duringCovid
```

```python
total_passenger_duringCovid = filtered_data_duringCovid['During Covid'].count()
total_passenger_duringCovid
```

```python
top20_passenger_duringCovid = filtered_data_top20_duringCovid['During Covid'].count()
top20_passenger_duringCovid
```

```python
top20_passenger_duringCovid_percentage = (top20_passenger_duringCovid/total_passenger_duringCovid)*100
top20_passenger_duringCovid_percentage
```
